## Requirement 3

A. Creation of a feature ("Customer_For") representing the number of days since customers started shopping at the store relative to the last recorded date (Recency).


```python
from datetime import datetime, date, timedelta

# Get today's date
today = datetime.now()

# Create an empty list
customer_for_list = []

for signup, recency in zip(df['Dt_Customer'], df['Recency']):
  # Get customer's last purchase date
  last_buy = today - timedelta(days = recency)
  # Subtract enrollment date from last purchase date
  customer_for_list.append((last_buy - signup).days)

# Create the new column 'Customer_For'
df['Customer_For'] = customer_for_list
```

B. Extraction of the age ("Age") of a customer based on the "Year_Birth" indicating the birth year of the respective individual.

```python
# Get current year
year = int(today.strftime('%Y'))

# Create the new column 'Age'
df['Age'] = year - df['Year_Birth']
```

Γ. Create another feature ("Spent") indicating the total amount spent by the customer across all categories within a two-year period.


```python
# Select 'MntWines', 'MntFruits', ..., 'MntGoldProds' columns to sum
df['Spent'] = df.iloc[:,9:15].sum(axis=1)
```

Δ. Create a feature "Children" to indicate the total number of children in a household, including both children and teenagers.

```python
# Select 'Kidhome' and 'Teenhome' columns to sum
df['Children'] = df.iloc[:,5:7].sum(axis=1)
```

Ε. To gain further clarity on the household, create a feature labeled "Family_Size" indicating the total number of individuals in a household.

```python
# Single = 0 | Together = 1 | Married = 2 | Divorced = 3 | Widow = 4

# Create an empty list
fam_size_list = []

# Traverse 'Marital_Status' and 'Children' columns,
# calculate the amount of people in each household
# and append it into the list 'fam_size_list'
for status, childs in zip(df['Marital_Status'], df['Children']):
  if status == 1 or status == 2:
    fam_size_list.append(childs + 2)
  elif status == 0 or status == 3 or status == 4:
    fam_size_list.append(childs + 1)
  else:
    fam_size_list.append('')

# Create the new column 'Family_Size'
df['Family_Size'] = fam_size_list
```

Στ. Create a feature "Is_Parent" indicating whether a customer is a parent or not.

```python
# Parent = 1 | Not Parent = 0

# Create an empty list
parent_list = []

# Traverse 'Children' column,
# check whether customer is parent or not
# and append it into the list 'parent_list'
for childs in df['Children']:
  if childs > 0:
    parent_list.append(1)
  else:
    parent_list.append(0)

# Create the new column 'Is_Parent'
df['Is_Parent'] = parent_list
```

Ζ. Create another feature "Living_With" using the "Marital_Status" to extract the living arrangement of couples. Specifically, this feature should have two values, "Partner" and "Alone".


```python
# Single = 0 | Together = 1 | Married = 2 | Divorced = 3 | Widow = 4

# Create an empty list
live_with_list = []

# Traverse 'Marital_Status' column and append
# 'Partner' or 'Alone' into the list 'live_with_list'
for status in df['Marital_Status']:
  if status == 1 or status == 2:
    live_with_list.append('Partner')
  elif status == 0 or status == 3 or status == 4:
    live_with_list.append('Alone')
  else:
    live_with_list.append('')

# Create the new column 'Living_With'
df['Living_With'] = live_with_list
```

Η. Create the column "Age Group" using the "Age" column, which groups ages into the following values: "21-30", "31-40", "41-50", "51-60", "61-70", "71-80", and ">80".

```python
# Create an empty list
age_grp_list = []

# Traverse 'Age' column and append
# the corresponding age group into the list 'age_grp_list'
for age in df['Age']:
  if age > 80:
    age_grp_list.append('>80')
  elif age >= 71 and age <= 80:
    age_grp_list.append('71-80')
  elif age >= 61 and age <= 70:
    age_grp_list.append('61-70')
  elif age >= 51 and age <= 60:
    age_grp_list.append('51-60')
  elif age >= 41 and age <= 50:
    age_grp_list.append('41-50')
  elif age >= 31 and age <= 40:
    age_grp_list.append('31-40')
  elif age >= 21 and age <= 30:
    age_grp_list.append('21-30')
  else:
    age_grp_list.append('')

# Create the new column 'Age Group'
df['Age Group'] = age_grp_list
```

## Requirement 4

Check for outliers in the features and remove them from the data.

```python
# Make a copy of original df
newdf = df.copy()

# Columns to be checked for outliers
cols_list = ['Income', 'Recency', 'NumDealsPurchases', 'NumWebPurchases',
             'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth',
             'Customer_For', 'Spent']

# For every column in 'cols_list'
for col in cols_list:
  # Check whether or not it's an outlier
  newdf['Outlier'] = abs(newdf[col] - newdf[col].mean()) > 1.96*newdf[col].std()
  # If it's an outlier, then drop the whole record
  newdf.drop(newdf[newdf['Outlier'] == True].index, inplace=True)

# Column 'Outlier' no longer needed
newdf.drop(['Outlier'], axis=1, inplace=True)

newdf
```

## Requirement 7 (PCA)

We will use the `LabelEncoder()` to convert the features `Education`, `Marital_Status` and `Living_With` into numeric data. Then, if necessary, each value of each category can be transformed into a new column where it will be assigned the value 0 or 1, i.e., True/False (this process is called `One-Hot Encoding` and can be applied as indicated in the [`06Label Encoding in Python.ipynb`](https://eclass.uoa.gr/modules/document/file.php/D424/%CE%A6%CF%81%CE%BF%CE%BD%CF%84%CE%B9%CF%83%CF%84%CE%AE%CF%81%CE%B9%CE%B1/06Label%20Encoding%20in%20Python.ipynb) notebook).

```python
from sklearn.preprocessing import LabelEncoder

# Creating instance of LabelEncoder
labelencoder = LabelEncoder()

# Convert into numerical values using LabelEncoder
newdf['Education'] = labelencoder.fit_transform(newdf['Education'])
newdf['Marital_Status'] = labelencoder.fit_transform(newdf['Marital_Status'])
newdf['Living_With'] = labelencoder.fit_transform(newdf['Living_With'])
```

Delete the columns related to offers and promotions. Also, remove the column `Dt_Customer` as it contains dates, `Age Group` which is a string, `ID` which is simply the identifier of each customer, `Year_Birth` and `Recency` as they are essentially duplicates of `Age` and `Customer_For` respectively, `Z_CostContact` and `Z_Revenue` since they contain only the numbers `3` and `11` respectively.

```python
# Make a copy of our df
final_df = newdf.copy()

final_df.drop(['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4',
               'AcceptedCmp5', 'Complain', 'Response', 'Dt_Customer',
               'Age Group', 'ID', 'Year_Birth', 'Recency', 'Z_CostContact',
               'Z_Revenue'],
              axis=1, inplace=True)
```

We perform `Standardization` so that each column/feature/variable has a mean `μ = 0` and a standard deviation `σ = 1`. Then, we use the `Principal Component Analysis (PCA)` method to reduce the dimensions to `n_components=3`. Finally, we plot the (three-dimensional) projection of the result.


```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits import mplot3d

# Scale the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(final_df)

# Use PCA to reduce dimensions to n_components=3
pca = PCA(n_components=3)
pca_data = pca.fit_transform(scaled_features)

# Store PCA results in a dataframe
df_pca = pd.DataFrame(pca_data, columns=['PC1', 'PC2', 'PC3'])

# Make the 3D projection
fig = plt.figure(figsize=(15, 15))
ax = plt.axes(projection='3d')
ax.scatter3D(df_pca['PC1'], df_pca['PC2'], df_pca['PC3'], s=40)
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('3D Projection')
plt.show()
```

## Requirement 8 (Clustering)

* The ELBOW method is used to determine the number of clusters that should be formed.

```python
from sklearn.cluster import KMeans

# Calculate the score for different K-mean values
sse = []
for k in range(1, 11):
  kmeans = KMeans(n_clusters=k, n_init='auto')
  kmeans.fit(pca_data)
  sse.append(kmeans.inertia_)

plt.plot(range(1, 11), sse)
plt.xticks(range(1, 11))
plt.title("ELBOW Method")
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

* Clustering using Agglomerative and K-Means hierarchical clustering.

```python
# Number of clusters to be used based on ELBOW Method
num_clusters = 4
```

```python
# K-Means
kmeans = KMeans(n_clusters=num_clusters, n_init='auto')
labels_kmeans = kmeans.fit_predict(pca_data)
df_pca['Cluster-KMeans'] = labels_kmeans
```

```python
# Agglomerative
from sklearn.cluster import AgglomerativeClustering
agglom = AgglomerativeClustering(n_clusters=num_clusters)
labels_aggl = agglom.fit_predict(pca_data)
df_pca['Cluster-Agglomerative'] = labels_aggl
```

* Displaying the formed clusters via a plot (e.g., scatter plot).

```python
# Make the 3D projection with 'num_clusters' clusters,
# each having different color
fig = plt.figure(figsize=(15, 15))
ax = plt.axes(projection='3d')
for i in range(num_clusters):
    ax.scatter3D(df_pca[df_pca['Cluster-KMeans'] == i]['PC1'],
                 df_pca[df_pca['Cluster-KMeans'] == i]['PC2'],
                 df_pca[df_pca['Cluster-KMeans'] == i]['PC3'],
                 s=40,
                 label=f'Cluster {i+1}')
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title(f'3D Projection with {num_clusters} Clusters (K-Means)')
ax.legend()
plt.show()
```
